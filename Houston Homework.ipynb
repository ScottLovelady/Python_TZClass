```python
import pandas as pd
import json

with open('Timeseries_Consumption.json', 'r') as f:
    data=json.loads(f.read())
objs=list()
data=data['account']['timeseriesdataidr']
for i in range(len(data)):
    objs.append(pd.DataFrame(data[i]['reads']))
data=pd.concat(objs)
data['t']=pd.to_datetime(data['t'])
data
```
